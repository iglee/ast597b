# PROBLEM 2.1 - 2.2

```python
import pandas as pd
import sqlite3
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
con = sqlite3.connect("dered.db")
```

```python
test = pd.read_sql("""SELECT * FROM data""",con)
```

```python
x = np.array(test['right_ascension'])
```

```python
y = np.array(test['declination'])
```

# PROBLEM 2.3

```python
plt.scatter(x,y,marker='.', s = 0.001)
plt.ylim([-2.5,4])
plt.xlim([225,235])
plt.xlabel('ra')
plt.ylabel('dec')
plt.title('ra vs dec')
plt.show()
```

# PROBLEM 2.4.1 -python numpy method

```python
x1=np.where(x>=228.5)
x2=np.where(x<=229.5)
ind_x=np.intersect1d(x1, x2)
```

```python
y1=np.where(y>=(-0.5))
y2=np.where(y<=0.5)
ind_y=np.intersect1d(y1, y2)
```

```python
ind = np.intersect1d(ind_x, ind_y)
```

```python
ra_sample=x[ind]
dec_sample=y[ind]
plt.scatter(ra_sample,dec_sample,marker='.', s = 0.1)
plt.ylim([-0.5,0.5])
plt.xlim([228.5,229.5])
plt.xlabel('ra')
plt.ylabel('dec')
plt.title('ra vs dec')
plt.show()
```

# PROBLEM 2.4.2 -SQL query method

```python
test1 = pd.read_sql("""SELECT * FROM data
WHERE right_ascension BETWEEN 228.5 AND 229.5
AND declination BETWEEN -0.5 AND 0.5;
""",con)
```

```python
x2 = np.array(test1['right_ascension'])
y2 = np.array(test1['declination'])
plt.scatter(x2,y2,marker='.', s = 0.1)
plt.ylim([-0.5,0.5])
plt.xlim([228.5,229.5])
plt.xlabel('ra')
plt.ylabel('dec')
plt.title('ra vs dec')
plt.show()
```

```python
test2 = pd.read_sql("""SELECT * FROM data
WHERE right_ascension BETWEEN 230 AND 231
AND declination BETWEEN -0.5 AND 0.5;
""",con)
```

```python
G = np.array(test2['dered_g'])
R = np.array(test2['dered_r'])
GR = G-R

G1 = np.array(test1['dered_g'])
R1 = np.array(test1['dered_r'])
GR1 = G1-R1
```

```python
plt.scatter(GR,G,marker='.', s = 0.001)
plt.ylim([10,30])
plt.xlim([-10,10])
plt.gca().invert_yaxis()
plt.xlabel('g-r')
plt.ylabel('g')
plt.title('g-r vs g')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(GR,G,marker='.', s = 0.1)
plt.ylim([16,22])
plt.xlim([-.5,2])
plt.gca().invert_yaxis()
plt.xlabel('g-r')
plt.ylabel('g')
plt.title('g-r vs g')

plt.subplot(1,2,2)
plt.scatter(GR1,G1,marker='.', s = 0.1)
plt.ylim([16,22])
plt.xlim([-.5,2])
plt.gca().invert_yaxis()
plt.xlabel('g-r')
plt.ylabel('g')
plt.title('g-r vs g')
plt.show()
```
