# Part 1 - Write a query (either SQL or for LSD) to select stars satisfying the g−r<0.4 color selection criterion

```python
from astropy.io import fits as pyfits
t1 = pyfits.open('stars1.fit')
t2 = pyfits.open('stars2.fit')
t3 = pyfits.open('stars3.fit')
t4 = pyfits.open('stars4.fit')
t5 = pyfits.open('stars5.fit')
t6 = pyfits.open('stars6.fit')
t7 = pyfits.open('stars7.fit')
t8 = pyfits.open('stars8.fit')
t11 = pyfits.open('stars11.fit')
t13 = pyfits.open('stars13.fit')
t14 = pyfits.open('stars14.fit')
```

```python
ra = []
dec = []
dered_r = []
```

```python
for t in (t1, t2, t3, t4, t5, t6, t7, t8, t11, t13, t14):
    for x in t[1].data['ra']:
        ra.append(x)
    for x in t[1].data['dec']:
        dec.append(x)
    for x in t[1].data['dered_r']:
        dered_r.append(x)
```

```python
ra = np.array(ra)
dec = np.array(dec)
x = np.array(dered_r)

ra_index = np.where((ra>=110)&(ra<=230))
dec_index = np.where((dec>=-5)&(dec<=60))
ind_p=np.intersect1d(ra_index, dec_index)
```

```python
ra = ra[ind_p]
dec = dec[ind_p]
x = x[ind_p]
```
